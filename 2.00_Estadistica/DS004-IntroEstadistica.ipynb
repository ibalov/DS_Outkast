### Media geométrica

```python
import scipy.stats

# Definimos un array de NumPy con valores
array = [1, 10, 100, 1000]

# Calculamos la MEDIA GEOMÉTRICA usando scipy.stats.gmean
media_geometrica = scipy.stats.gmean(array)
media_aritmetica = np.mean(array)

print(f"La media aritmética del array es: {media_aritmetica}")
print(f"La media geométrica del array es: {media_geometrica}")
```

### Media Armónica

```python
# Definimos un array de NumPy con valores
array = [1, 2, 4, 8]

# Calculamos la MEDIA ARMÓNICA usando scipy.stats.hmean
media_armonica = scipy.stats.hmean(array)
media_aritmetica = np.mean(array)

print(f"La media armónica del array es: {media_armonica}")
print(f"La media aritmética del array es: {media_aritmetica}")
```

### Comparativa de medias

```python
# Velocidades en km/h para el viaje de ida y vuelta
velocidades = np.array([60, 40])

# Calculamos la media aritmética con NumPy
media_aritmetica = np.mean(velocidades)
print(f"La media aritmética de las velocidades es: {media_aritmetica} km/h")

# Calculamos la media armónica con SciPy
media_armonica = scipy.stats.hmean(velocidades)
print(f"La media armónica de las velocidades es: {media_armonica} km/h")

# Supongamos que la distancia del viaje en una sola dirección es de 240 km
distancia = 240

# Tiempo gastado para cada segmento del viaje
tiempo_ida = distancia / velocidades[0]
tiempo_vuelta = distancia / velocidades[1]

# Tiempo total del viaje
tiempo_total = tiempo_ida + tiempo_vuelta

# Distancia total del viaje
distancia_total = distancia * 2

# Velocidad media real del viaje completo
velocidad_media_real = distancia_total / tiempo_total
print(f"La velocidad media real del viaje completo es: {velocidad_media_real} km/h")
```

### Media ponderada

```python
# Calificaciones en cada componente
calificaciones = np.array([80, 90, 70])

# Peso de cada componente en la calificación final
pesos = np.array([0.3, 0.5, 0.2])

# Calcular la media ponderada
media_ponderada = np.average(calificaciones, weights=pesos)

print(f"La calificación final del estudiante en el curso es: {media_ponderada}")
```

En estadística, la media aritmética es la media de tendencia central, mientras que la mediana es el valor que separa la mitad superior de un conjunto de datos de su mitad inferior

```python
# Definimos un array de NumPy con valores
array = np.array([1, 2, 4, 8])

# Calculamos las medias: armónica, aritmética y geométrica
media_armonica = scipy.stats.hmean(array)
media_aritmetica = np.mean(array)
media_geometrica = scipy.stats.gmean(array)

# Crear el gráfico
plt.figure(figsize=(10, 6))

# Dibujar los valores del array como puntos
plt.scatter(range(len(array)), array, label='Valores individuales', color='blue')

# Dibujar la línea horizontal para la media armónica
plt.axhline(y=media_armonica, color='r', linestyle='--', label=f'Media armónica = {media_armonica:.2f}')

# Dibujar la línea horizontal para la media aritmética
plt.axhline(y=media_aritmetica, color='g', linestyle='--', label=f'Media aritmética = {media_aritmetica:.2f}')

# Dibujar la línea horizontal para la media geométrica
plt.axhline(y=media_geometrica, color='purple', linestyle='--', label=f'Media geométrica = {media_geometrica:.2f}')

# Añadir etiquetas y leyenda
plt.xlabel('Índice')
plt.ylabel('Valor')
plt.title('Comparación de medias con valores individuales')
plt.legend()

# Mostrar el gráfico
plt.show()
```

### La Mediana

- Definición: Es eñ vañpr dentral de un conjunto de datos ordenados
- Calcular: Se debe ordenar ascendente o descendente, elije el valor del centro
- Aplicaciones: Para describir grandes conjuntos de datos donde los valores atípicaos pueden sesgar la media aritmética

```python
#Para nº impares
arrayImpar = np.array([1,2,3,4,5])
mediana = np.median(arrayImpar)
print(f"Mediana: {mediana}") #resultado
```

```python
#Para nº Pares
arrayPar = np.array([4,1,6,9])
np.sort(arrayPar)
mediana = np.median(arrayPar)
#Debemos sumar sus 2 pares centrales-> 4+6=10 para luego 10/2 =5
print(f"Mediana: {mediana}") #resultado
```

**La principal diferencia entre Media y Mediana es la sensibilidad ante los valores atípicos** 

```python
array = np.array([1,2,3,4,100])
print(f"Media con valor atípico: {np.mean(array)}")
print(f"Mediana con valor atípico: {np.median(array)}")
```
