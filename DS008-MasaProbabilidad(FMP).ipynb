La función `norm.pdf(x)` representa la función de densidad de probabilidad (PDF, por sus siglas en inglés) de una distribución normal estándar. Específicamente, esta función describe la probabilidad de que una variable aleatoria continua tome un valor específico \(x\). La forma matemática de esta función es:

$$
\text{norm.pdf}(x) = \frac{\exp\left(-\frac{x^2}{2}\right)}{\sqrt{2\pi}}
$$

La curva resultante, conocida como la "campana de Gauss", es simétrica alrededor del valor \(x = 0\) y tiene una forma de campana. Los valores más altos de \(x\) (tanto positivos como negativos) resultarán en valores más pequeños de la PDF, lo que indica que es menos probable que la variable aleatoria tome esos valores.

```python
import matplotlib.pyplot as plt

# Datos
numeros = ['1', '2', '3', '4', '5', '6'] #los n1 de los dados
probabilidades = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6] #probabilidades de sacar el número

# Creación del gráfico
plt.bar(numeros, probabilidades)
plt.xlabel('Número del dado')
plt.ylabel('Probabilidad')
plt.title('Probabilidad de cada número en un lanzamiento de dado')
plt.ylim(0, 0.2)  # Establecer límites del eje y
plt.show()
```

### Usando la función de Masa Probabilidad

```python
import numpy as np
import matplotlib.pyplot as plt

def funcion_masa_probabilidad(x, resultados):
    """
    Retorna la probabilidad de un evento x dada una lista de posibles resultados 'resultados'.
    En este caso, simplemente dividimos 1 entre el número de posibles resultados.
    """
    if x in resultados:
        return 1 / len(resultados)
    else:
        return 0

# Definimos los posibles resultados de un lanzamiento de dado
resultados = [1, 2, 3, 4, 5, 6]

# Calculamos las probabilidades para cada número del dado usando la FMP
probabilidades = [funcion_masa_probabilidad(x, resultados) for x in resultados]

# Creación del gráfico
plt.bar([str(x) for x in resultados], probabilidades)
plt.xlabel('Número del dado')
plt.ylabel('Probabilidad')
plt.title('Probabilidad de cada número en un lanzamiento de dado')
plt.ylim(0, 0.2)
plt.show()
```

### 2.1 Tiene que cumplir 3 propiedades

- **No Negatividad: 0<=P(X=k)<=1: Los valores no pueden ser negativos. P(X=x)=>1/6=>0**
- **Suma de Probabilidades=1: La suma de las probabilidades es igual a 1. Por ejemplo: P(X=1)+P(X=2)+P(X=3)+P(X=4)+P(X=5)+P(X=6)**
- **Valores mutuamente excluyentes: No pueden ocurrir simultáneamente. Por ejemplo, en el lanzamiento de un dado no se puede obtener los valores de 1 y 2**

# 2.2 Función de Masa de Probabilidad (FMP)

Para una función de masa de probabilidad (FMP) de una variable aleatoria discreta \( X \), no existe una "fórmula genérica" única. Esto se debe a que la FMP depende específicamente de la naturaleza del experimento o proceso aleatorio que se está modelando. No obstante, existen dos propiedades generales que toda FMP debe satisfacer:

1. $$ 0 \leq P(X = k) \leq 1 $$ para cada valor posible \( k \).  
   Esto indica que las probabilidades siempre se encuentran entre 0 y 1.

2. $$ \sum_{\text{todos } k} P(X = k) = 1 $$.  
   Esto señala que la suma de las probabilidades de todos los posibles valores que \( X \) puede tomar es 1.

La "fórmula" específica para \( P(X = k) \) dependerá del problema en cuestión. Veamos algunos ejemplos:

- Si estás trabajando con un dado justo, la FMP es:  
  $$ P(X = k) = \frac{1}{6} $$ para \( k \) en \( \{1, 2, 3, 4, 5, 6\} \).

- Si estás modelando el número de éxitos en \( n \) intentos en un experimento de Bernoulli, la FMP está dada por la **distribución binomial**:  
  $$ P(X = k) = \binom{n}{k} p^k (1 - p)^{n-k} $$  
  donde \( p \) es la probabilidad de éxito en un solo intento.

Existen muchas otras distribuciones discretas (geométrica, poisson, hipergeométrica, etc.) y cada una tiene su propia fórmula para la FMP. Por tanto, para calcular la probabilidad en una FMP, es esencial conocer la naturaleza de la distribución que se está manejando y aplicar la fórmula correspondiente.

<hr>

**Distribución Binomial:**

Supón que realizas un experimento de Bernoulli \( n \) veces y estás interesado en la cantidad de veces que ocurre un éxito. Si cada intento del experimento es independiente, entonces la cantidad de éxitos sigue una distribución binomial. La probabilidad de obtener exactamente \( k \) éxitos en \( n \) intentos es:

$$
P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}
$$

Donde:
- $$ \binom{n}{k} $$ es el **coeficiente binomial**, que indica el número de maneras en que puedes elegir \( k \) éxitos de \( n \) intentos. El coeficiente binomial calcula todas las posibles combinaciones de que se pueda obtener "k" exitos. Con "n" lanzamientos/intentos.
El CB se calcula:$$
\binom{n}{k} = \frac{n!}{k!(n-k)!}
$$

- $$ p^k $$ es la probabilidad de obtener \( k \) éxitos.
- $$ (1-p)^{n-k} $$ es la probabilidad de obtener \( n-k \) fracasos.

**2. Práctica con la fórmula matemática:**

Ejemplo: Supón que lanzas una moneda justa (donde \( p = 0.5 \)) 10 veces. ¿Cuál es la probabilidad de obtener exactamente 3 caras?

Usando la fórmula de la distribución binomial:

$$
P(X = 3) = \binom{10}{3} (0.5)^3 (1 - 0.5)^{10-3}
$$

$$
P(X = 3) = 120 \times 0.125 \times 0.5^7
$$

$$
P(X = 3) \approx 0.117
$$

Así que la probabilidad de obtener exactamente 3 caras en 10 lanzamientos es aproximadamente 0.117 o 11.7%.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom #funcionalidades para la distribucion binomial

# Parámetros
n = 10 #nº intentos
p = 0.5 #Probabilidad
x = np.arange(0, n+1) #generamos una secuencia de nº desde 0 hasta n+1 np.arrange(start,stop,step)

# Probabilidades usando distribución binomial
probabilidades = binom.pmf(x, n, p)  

# Graficar
plt.bar(x, probabilidades)
plt.xlabel('Número de éxitos')
plt.ylabel('Probabilidad')
plt.title('Distribución Binomial para n=10 y p=0.5')
plt.show()
```

```python
import pandas as pd

# Crea un DataFrame con los datos
df = pd.DataFrame({
    'Número de éxitos': x,
    'Probabilidad': probabilidades
})

# Muestra el DataFrame
print(df)
```
